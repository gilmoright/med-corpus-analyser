```python
import os
import re
import json
```

```python
original_file = "/s/ls4/users/zsaveliy/RE/RelationExtraction/data/otzovik/otzovik_for_rdrs_full.json"
parsed_file = "/s/ls4/users/grartem/SAG_MED/med-corpus-analyser/data/raw/ner_800k_jsonl/ner_800k.jsonl"
```

```python
with open(original_file, "r") as f:
    original_data = json.load(f)
```

```python
urls = []
with open(parsed_file, "r") as f:
    l_i = 0
    for line in f:
        data = json.loads(line)
        fixed = re.sub("[^а-яА-Я]", "", original_data[l_i]["raw"])
        fixed2 = re.sub("[^а-яА-Я]", "", data["text"])
        #assert len(fixed)>10
        if fixed2!=fixed:
            print(l_i)
        urls.append(original_data[l_i]["url"])
        l_i+=1
```

```python
tags_set = set()
with open(parsed_file, "r") as f:
    l_i = 0
    for line in f:
        data = json.loads(line)
        for k,v in data["entities"].items():
            tags_set.add(v["tag"][0])
            assert len(v["tag"])==1
            assert "MedFrom" not in v["tag"][0]
```

```python
# save terms for normalization
short_data = []
with open(parsed_file, "r") as f, open("/s/ls4/users/grartem/SAG_MED/med-corpus-analyser/data/raw/ner_800k_jsonl/short_ner_800k.jsonl", "w") as outf:
    l_i = 0
    for line in f:
        data = json.loads(line)
        data.pop("text")
        for entity_i, entity in data["entities"] .items():
            assert len(entity["tag"])==1
            assert entity["entity_id"]==int(entity_i)
            entity.pop("entity_id")
            entity["tag"] = entity["tag"][0]
            entity.pop("spans")
        outf.write(json.dumps(data)+"\n")
```
